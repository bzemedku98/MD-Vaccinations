```python
pop = df[["Unnamed: 0", "Unnamed: 13"]]
pop
```

```python
pop = pop.rename(columns={"Unnamed: 0": "Counties", "Unnamed: 13": "Population"})
pop
```

```python
pop["Counties"] = pop["Counties"].str.strip()
pop
```

```python
pop = pop[pop["Counties"] != ""]
pop
```

```python
pop.to_csv("populations-md.csv", index=False, encoding="utf-8")
```
